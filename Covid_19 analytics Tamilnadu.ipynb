```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as ex

df = pd.read_csv('../Desktop/districtwise_statistics_of_covid_19_cases_in_tn_as_on_28_04_2021.csv',index_col=0)
df.head()
```

```python
### Rename column for conveniance
df.rename(columns={'Total Positive Cases as on 28.04.2021':'Positive cases','Discharged Cases as on 28.04.2021':'Discharged cases','Active Cases as on 28.04.2021':'Active cases','Death Cases as on 28.04.2021':'Death cases'}, inplace=True)
df.tail()
```

```python
### Drop 'Total' row
dropped_Total_row = df.drop(['Total'],axis=0)
dropped_Total_row.tail()
```

```python
### Plotting district wise positive cases
### Sort values
sorted_value = dropped_Total_row.sort_values(by=['Positive cases'],ascending=False)
### Generate a barplot
plt.figure(figsize=(12,25))
plot = sns.barplot(sorted_value['Positive cases'], sorted_value['District'])

for i, (value,name) in enumerate(zip(sorted_value['Positive cases'],sorted_value['District'])):
    plot.text(value,i-0.05,f'{value:,.0f}',size=10)
plt.show()
```

```python
### Plotting district wise positive cases and recovered patients

plt.figure(figsize=(15,25))
positive = sns.barplot(sorted_value['Positive cases'], sorted_value['District'],color='red',label='Positive cases')
recovered = sns.barplot(sorted_value['Discharged cases'], sorted_value['District'],color='green',label='Recovered')

for i,(value,name) in enumerate(zip(sorted_value['Positive cases'],sorted_value['District'])):
    positive.text(value,i-0.05,f'{value:,.0f}',size=9)
    
for i,(value,name) in enumerate(zip(sorted_value['Discharged cases'],sorted_value['District'])):
    recovered.text(value,i-0.05,f'{value:,.0f}',size=9)
    
plt.legend(loc=4)
plt.show()
```
